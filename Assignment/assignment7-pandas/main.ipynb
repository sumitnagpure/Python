## Q1.

Create a DataFrame containing student information like Name, Age, Department, %Marks

```python
df=pd.DataFrame({'Name':['Rohit','Sunil','Sujita'],
                 'Age':[21,25,31],
                 'Department':['CSE','Civil','Mech'],
                 'Marks':[91,89,92]})
print(df)
```

a.Display column names of the DF 

```python
print(df.loc[:-1,:])
print(df.columns)
```

 b.Show min/max and average marks  

```python
print("Min Marks: ",df.Marks.min())
print("Max Marks: ",df.Marks.max())
print("Avg Marks: ",df.Marks.mean())
print('\nagg()')
print(df.Marks.agg(['min','max','mean']))
```

c.Get unique values and student count for each Department

```python
print('Unique values:',df['Department'].unique())
print('\n',df['Department'].value_counts())
```

## Q2. Read csv file solar.csv  

```python
solar=pd.read_csv('solar.csv')
solar
```

a. Display all the plants with capacity > 500 MW

```python
filtered=solar[solar['Installed Capacity (MW)']>500].dropna()
filtered
```

b. Display plant details for New York plant  

```python
# Idiomatic approach
solar[solar['State']=='New York']
```

```python
# .query()
filtered=solar.query("State=='New York'")
filtered
```

```python
#  .loc
filtered=solar.loc[solar['State']=='New York']
filtered
```

c. For all the plants display Average MW Per Plant and Generation (GWh)

```python
filtered=solar[['State','Average MW Per Plant','Generation (GWh)']]
filtered
```

d. Sort the plants by Generation (GWh) (ascending) and Installed Capacity (MW)(descending) 

```python
sorted=solar.sort_values(by=['Generation (GWh)','Installed Capacity (MW)'],ascending=[True,False])
sorted
```

e. List top 5 plants for their Generation 

```python
top5=solar.sort_values('Generation (GWh)',ascending=False)[:5]
top5
```

```python
targer_states=['California','Nevada','Arizona','Texas']
mask=solar['State'].isin(targer_states)
filtered=solar[mask]
filtered
```

## Q.3

Read csv gapminder2007.csv

```python
df_gap=pd.read_csv('gapminder2007.csv')
df_gap.head()
```

a.Display data for all Asian countries

```python
df_gap[df_gap['continent']=='Asia']
```

b.Display top and bottom 15 rows

```python
pd.concat([df_gap.head(15),df_gap.tail(15)])
```

c. Show all the rows where life expectancy is more than 50 years and less than 80 years

```python
df_gap[(df_gap['lifeExp']>50) & (df_gap['lifeExp']<80)]
```

d. Show data for India, America and France

```python
df_gap[df_gap['country'].isin(['India','United States','France'])]
```

e. Show data for all countries where name starts with ‘A’

```python
df_gap[df_gap['country'].str.startswith('A')]
```

f. Show min/max and mean GDP and Population for each continent

```python
summ=df_gap.groupby(by='continent').agg({'pop':['min','max','mean'],'gdpPercap':['min','max','mean']})
summ
```

g. Show population and life expectancy for all countries 

```python
df_gap[['country','pop','lifeExp']]
```

h.Sort the data by country names (ascending) and GDP (descending)

```python
df_gap.sort_values(by=['country','gdpPercap'],ascending=[True,False])
```

i.Find top 20 most populous countries

```python
df_gap.sort_values(by='pop',ascending=False)[:20]
```

j.Delete all rows for a specific country

```python
country_to_drop='Iran'
index_to_drop=df_gap[df_gap['country']==country_to_drop].index
new_df=df_gap.drop(index=index_to_drop,axis=0)
'Iran' in new_df['country'].values
```

k.Change all column names to title case

```python
new_df.columns=df_gap.columns.str.title()
new_df.head()
```

l.Change column names: pop to Population, gdpPercap to GDP_Percap

```python
new_df=df_gap.rename(columns={'pop':'Population',
                              'gdpPercap':'GDP_Percap' })
new_df.head()
```

m.Increase Life Expectancy by 2 years for all African countries

```python
new_df=df_gap.copy()
mask=new_df['continent']=='Africa'
new_df.loc[mask,'lifeExp']=new_df.loc[mask,'lifeExp']+2
print("Old \n",df_gap[2:4])
print("New \n",new_df[2:4])
```

n.Find all the rows that contain NA values

```python
new_df=df_gap.copy()
new_df.loc[len(df_gap)]=['Atlantis',12.0,'Ocenia',85.0,float('nan')]
mask=new_df.isna().any(axis=1)
new_df[mask]
```

o.Fill NA values with 0 for numeric columns

```python
new_df=new_df.fillna(0)
# idiomatic approach
print(new_df.iloc[-1])
```

```python
new_df=df_gap.copy()
new_df.loc[len(df_gap)]=['Atlantis',12.0,'Ocenia',85.0,float('nan')]
new_df.iloc[-1]
```
