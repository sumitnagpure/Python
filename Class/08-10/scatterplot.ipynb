```python
# CODE TO RUN FOR SCATTERPLOT

import pandas as pd             # Import the pandas library for data manipulation.
import seaborn as sns           # Import the seaborn library for statistical plotting (aliased as sns).
import matplotlib.pyplot as plt # Import the matplotlib plotting library for plot control (aliased as plt).

# --- 1. DATA PREPARATION ---
# Load a sample dataset built into seaborn (e.g., 'tips').
data = sns.load_dataset('tips')

# --- 2. PLOT SETUP (Matplotlib) ---
# Create a figure and axes object, explicitly setting the plot size to 8 inches wide by 5 inches tall.
plt.figure(figsize=(8, 5))

# --- 3. CREATE THE SCATTERPLOT (Seaborn) ---
# Call the core function to generate the Scatterplot.
sns.scatterplot(
    data=data,              # Specify the DataFrame to use.
    x='total_bill',         # Specify the continuous variable for the horizontal axis (X-axis).
    y='tip',                # Specify the continuous variable for the vertical axis (Y-axis).
    hue='time',             # (Optional) Color the points based on a third, categorical variable ('Lunch' or 'Dinner').
    size='size',            # (Optional) Change the size of the points based on a fourth, numeric variable (party size).
    alpha=0.7               # (Optional) Set transparency for overlapping points.
)

# --- 4. FINALIZE AND DISPLAY (Matplotlib) ---
# Set the title of the plot.
plt.title('Total Bill vs. Tip Amount, Colored by Time')

# Add a grid to help read the values.
plt.grid(True, linestyle='--', alpha=0.5)

# Command to display the plot (essential in notebooks).
plt.show()
```
