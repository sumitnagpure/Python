# KDE Plot

```python
# CODE TO RUN FOR KDE PLOT (Kernel Density Estimate)

import pandas as pd             # Import the pandas library for data manipulation (using DataFrames).
import seaborn as sns           # Import the seaborn library for statistical plotting (aliased as sns).
import matplotlib.pyplot as plt # Import the matplotlib plotting library for plot control (aliased as plt).

# --- 1. DATA PREPARATION ---
# Load a sample dataset built into seaborn for demonstration.
data = sns.load_dataset('titanic')

# --- 2. PLOT SETUP (Matplotlib) ---
# Create a figure and axes object, explicitly setting the plot size to 8 inches wide by 5 inches tall.
plt.figure(figsize=(8, 5))

# --- 3. CREATE THE KDE PLOT (Seaborn) ---
# Call the core function to generate the Kernel Density Estimate plot.
sns.kdeplot(
    data=data,              # Specify the DataFrame to use.
    x='age',                # Specify the column whose distribution we want to plot (continuous variable).
    fill=True,              # Argument to shade the area under the density curve.
    bw_adjust=0.5,          # Adjusts the smoothness of the curve (Bandwidth). 0.5 makes it less smooth/more detailed.
    linewidth=2             # Sets the thickness of the curve line.
)

# --- 4. FINALIZE AND DISPLAY (Matplotlib) ---
# Set the title of the plot.
plt.title('KDE Plot of Passenger Ages')

# Set the label for the x-axis.
plt.xlabel('Age')

# Set the label for the y-axis (KDE plots show Density, not Counts).
plt.ylabel('Density')

# Command to display the plot (essential in notebooks).
plt.show()
```
