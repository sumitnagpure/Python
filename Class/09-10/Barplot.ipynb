# BarPlot

```python
import plotly.express as px
import pandas as pd
```

```python
df=pd.read_csv('gapminder2007.csv')
```

```python
px.bar(df,x='country',y='lifeExp',
       color='country')
```

```python
df=px.data.gapminder()
df
```

```python
fig=px.bar(df,x='country',y='lifeExp',
       color='country',
       animation_frame='year',  # Correct argument for animation time-step
       animation_group='country' # Groups the bars to track them across frames
       )
fig.show()
```

```python
fig = px.bar(
    df,
    x='country',             # The primary category (Country)
    y='pop',                 # The value to plot (Population)
    color='continent',       # Color the bars by Continent
    
    # These two arguments "stack the years on country" for the animation
    animation_frame='year',  # Creates a slide/frame for each year (1952, 1957, etc.)
    animation_group='country', # Ensures the same country bar is tracked across frames
    
    # Optional: Initial setup
    height=600,
    title="Population Growth by Country Over Time"
)

fig.show()
```

```python
fig = px.bar( df,x='country', y='pop', color='continent',animation_frame='year', animation_group='country', 
             groupmode='stack')
```
